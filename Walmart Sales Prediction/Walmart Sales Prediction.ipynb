```python
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
x = df_fl.drop(columns=['Sales(In ThousandDollars)'])
y = df_fl['Sales(In ThousandDollars)']
```

```python
x = pd.get_dummies(x, dtype=int, drop_first= True)
x.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import r2_score
```

```python
sr = StackingRegressor(estimators=[('xgboost',XGBRegressor())], final_estimator=GradientBoostingRegressor(), cv=5)
model = sr.fit(xtrain, ytrain)
ypred = model.predict(xtest)
r2_score(ytest,ypred)
```
